2. A town discharges 14 million litres per day of sewage at a temperature of 23°C into a
river having flow of 1.7 m3
/sec and water temperature of 20°C. BOD at 20°C for the
wastewater is 160 mg/L and kd is 0.23 per day. If kr is 0.26 per day. Assume the stream is
92% saturated with oxygen before the sewage addition and its BOD5 content is 3 mg/L.
The river flow velocity as 0.12 m/s. The DO saturation concentration can be calculated
using the following formula (T = temperature (oC)):

    DO saturation= 14.59-0.3955T+0.0072T^2-00000619T^3


a. Plot the profile of DO with distance (or time).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

Q_wastewater = 14*10^6
T_wastewater = 23
Q_river = 1.7
T_river = 20
BOD_wastewater_20C = 160
kd = 0.23
kr = 0.26
BOD5_river = 3
river_velocity = 0.12
```

```python

def calculate_DO_saturation(T):
    return 14.59 - 0.3955 * T + 0.0072 * T**2 - 0.0000619 * T**3

DO_source = calculate_DO_saturation(T_wastewater)
BOD_source = BOD_wastewater_20C * (1 - np.exp(-kd))

distance = np.arange(0, 2000, 10)

DO_concentration = np.zeros_like(distance)

for i, d in enumerate(distance):
    time = d / river_velocity
    DO_concentration[i] = DO_source + (BOD_source - BOD5_river) * (1 - np.exp(-kd * time)) - (
                DO_source - calculate_DO_saturation(T_river)) * np.exp(-kr * time)
```

```python



# Plot the DO profile
plt.figure(figsize=(10, 6))
plt.plot(distance, DO_concentration)
plt.xlabel('Distance (meters)')
plt.ylabel('Dissolved Oxygen (mg/L)')
plt.title('Dissolved Oxygen Profile along the River')
plt.grid(True)
plt.show()


```

b. Determine the critical oxygen deficit and the time and distance at which it occurs.

```python

critical_oxygen_deficit = min(DO_concentration) - calculate_DO_saturation(T_river)
index_critical_deficit = np.argmin(DO_concentration)
time_critical_deficit = distance[index_critical_deficit] / river_velocity

print("Critical Oxygen Deficit: {:.2f} mg/L".format(critical_oxygen_deficit))
print("Time at Critical Oxygen Deficit: {:.2f} days".format(time_critical_deficit))
print("Distance at Critical Oxygen Deficit: {:.2f} meters".format(distance[index_critical_deficit]))


```

c. If the value of kd is doubled, how would it affect the DO sag curve? Show your
answer on the same DO profile you plot in part (a) above.


```python

kd_doubled = 2 * kd

DO_concentration_doubled_kd = np.zeros_like(distance)

for i, d in enumerate(distance):
    time = d / river_velocity
    DO_concentration_doubled_kd[i] = DO_source + (BOD_source - BOD5_river) * (1 - np.exp(-kd_doubled * time)) - (
                DO_source - calculate_DO_saturation(T_river)) * np.exp(-kr * time)

plt.figure(figsize=(10, 6))
plt.plot(distance, DO_concentration, label='Original kd')
plt.plot(distance, DO_concentration_doubled_kd, label='Doubled kd', linestyle='--')
plt.xlabel('Distance (meters)')
plt.ylabel('Dissolved Oxygen (mg/L)')
plt.title('Dissolved Oxygen Profile along the River')
plt.grid(True)
plt.legend()
plt.show()


```

```python

kr_increased = 1.5 * kr

DO_concentration_increased_kr = np.zeros_like(distance)

for i, d in enumerate(distance):
    time = d / river_velocity
    DO_concentration_increased_kr[i] = DO_source + (BOD_source - BOD5_river) * (1 - np.exp(-kd * time)) - (
                DO_source - calculate_DO_saturation(T_river)) * np.exp(-kr_increased * time)

plt.figure(figsize=(10, 6))
plt.plot(distance, DO_concentration, label='Original kr')
plt.plot(distance, DO_concentration_increased_kr, label='Increased kr', linestyle='--')
plt.xlabel('Distance (meters)')
plt.ylabel('Dissolved Oxygen (mg/L)')
plt.title('Dissolved Oxygen Profile along the River')
plt.grid(True)
plt.legend()
plt.show()


```
