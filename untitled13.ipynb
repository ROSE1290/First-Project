2. A town discharges 14 million litres per day of sewage at a temperature of 23°C into a
river having flow of 1.7 m3
/sec and water temperature of 20°C. BOD at 20°C for the
wastewater is 160 mg/L and kd is 0.23 per day. If kr is 0.26 per day. Assume the stream is
92% saturated with oxygen before the sewage addition and its BOD5 content is 3 mg/L.
The river flow velocity as 0.12 m/s. The DO saturation concentration can be calculated
using the following formula (T = temperature (oC)):

    DO saturation= 14.59-0.3955T+0.0072T^2-00000619T^3


a. Plot the profile of DO with distance (or time).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#given data
sewage_flow_rate = 14*10^6
sewage_temperature = 23
river_flow_rate = 1.7
river_temperature = 20
river_BOD5 = 3
sewage_BOD = 160
kd = 0.23
kr = 0.26
river_flow_velocity = 0.12

#constant
A = 14.59
B = -0.3955
C = 0.0072
D = -0.0000619
```

```python

# sewage and rivwr water do saturation
def calculate_DO_saturation(T):
    return A + B * T + C * T**2 + D * T**3

sewage_DO_saturation = calculate_DO_saturation(sewage_temperature)
river_DO_saturation = calculate_DO_saturation(river_temperature)


time_or_distance = []
DO_profile = []

for t in range(0, 100, 25):
    sewage_DO_change = (sewage_BOD - river_BOD5) * (1 - (sewage_DO_saturation / calculate_DO_saturation(sewage_temperature)))
    river_DO_change = (sewage_BOD - river_BOD5) * (1 - (river_DO_saturation / calculate_DO_saturation(river_temperature)))
    total_change = (sewage_flow_rate / (river_flow_rate * 1000)) * (sewage_DO_change - river_DO_change)

    time_or_distance.append(t)
    DO_profile.append(river_DO_saturation + total_change)
```

```python



plt.plot(time_or_distance, DO_profile)
plt.xlabel('Time (seconds)')
plt.ylabel('Dissolved Oxygen (mg/L)')
plt.title('Dissolved Oxygen Profile Over Time')
plt.grid(True)
plt.show()

```

b. Determine the critical oxygen deficit and the time and distance at which it occurs.

```python
# Calculate COD and corresponding time and distance
COD = min(DO_profile)
COD_index = DO_profile.index(COD)
COD_time = time_or_distance[COD_index]
COD_distance = COD_time * river_flow_velocity

print("Critical Oxygen Deficit (COD):", COD, "mg/L")
print("Time at which COD occurs:", COD_time, "seconds")
print("Distance at which COD occurs:", COD_distance, "meters")

```

c. If the value of kd is doubled, how would it affect the DO sag curve? Show your
answer on the same DO profile you plot in part (a) above.


```python

# Double kd value
kd *= 2

DO_profile_doubled_kd = []

for t in time_or_distance:
    sewage_DO_change = (sewage_BOD - river_BOD5) * (1 - (sewage_DO_saturation / calculate_DO_saturation(sewage_temperature)))
    river_DO_change = (sewage_BOD - river_BOD5) * (1 - (river_DO_saturation / calculate_DO_saturation(river_temperature)))

    total_change = (sewage_flow_rate / (river_flow_rate * 1000)) * (sewage_DO_change - river_DO_change) * 2

    DO_profile_doubled_kd.append(river_DO_saturation + total_change)
plt.plot(time_or_distance, DO_profile, label='Original DO Profile')
plt.plot(time_or_distance, DO_profile_doubled_kd, label='DO Profile with Doubled kd')
plt.xlabel('Time (seconds)')
plt.ylabel('Dissolved Oxygen (mg/L)')
plt.title('Dissolved Oxygen Profile Over Time')
plt.grid(True)
plt.legend()
plt.show()

```

d. If kr is increased by 50%, how would it affect the DO sag curve? Show your answer
on the same DO profile you plot in part (a) above

```python
# Increase the value of kr by 50%
kr *= 1.5
DO_profile_increased_kr = []

for t in time_or_distance:
    sewage_DO_change = (sewage_BOD - river_BOD5) * (1 - (sewage_DO_saturation / calculate_DO_saturation(sewage_temperature)))
    river_DO_change = (sewage_BOD - river_BOD5) * (1 - (river_DO_saturation / calculate_DO_saturation(river_temperature)))

    total_change = (sewage_flow_rate / (river_flow_rate * 1000)) * (sewage_DO_change - river_DO_change)

    DO_profile_increased_kr.append(river_DO_saturation + total_change)

plt.plot(time_or_distance, DO_profile, label='Original DO Profile')
plt.plot(time_or_distance, DO_profile_doubled_kd, label='DO Profile with Doubled kd')
plt.plot(time_or_distance, DO_profile_increased_kr, label='DO Profile with Increased kr')
plt.xlabel('Time (seconds)')
plt.ylabel('Dissolved Oxygen (mg/L)')
plt.title('Dissolved Oxygen Profile Over Time')
plt.grid(True)
plt.legend()
plt.show()

```
