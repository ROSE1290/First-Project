2. A town discharges 14 million litres per day of sewage at a temperature of 23°C into a
river having flow of 1.7 m3
/sec and water temperature of 20°C. BOD at 20°C for the
wastewater is 160 mg/L and kd is 0.23 per day. If kr is 0.26 per day. Assume the stream is
92% saturated with oxygen before the sewage addition and its BOD5 content is 3 mg/L.
The river flow velocity as 0.12 m/s. The DO saturation concentration can be calculated
using the following formula (T = temperature (oC)):

    DO saturation= 14.59-0.3955T+0.0072T^2-00000619T^3


a. Plot the profile of DO with distance (or time).
b. Determine the critical oxygen deficit and the time and distance at which it occurs.
c. If the value of kd is doubled, how would it affect the DO sag curve? Show your
answer on the same DO profile you plot in part (a) above.
d. If kr is increased by 50%, how would it affect the DO sag curve? Show your answer
on the same DO profile you plot in part (a) above


```python

import numpy as np
import matplotlib.pyplot as plt

# Constants and data
Q_wastewater = 14e6  # Flow rate of wastewater in liters per day
T_wastewater = 23  # Temperature of wastewater in °C
Q_river = 1.7  # Flow rate of the river in m^3/sec
T_river = 20  # Temperature of the river in °C
BOD_wastewater_20C = 160  # BOD at 20°C for wastewater in mg/L
kd = 0.23  # Decay coefficient in per day
kr = 0.26  # Reaeration coefficient in per day
BOD5_river = 3  # BOD5 content of the river in mg/L
river_velocity = 0.12  # River flow velocity in m/s

# DO saturation concentration formula
def calculate_DO_saturation(T):
    return 14.59 - 0.3955 * T + 0.0072 * T**2 - 0.0000619 * T**3

# Create time values
time = np.linspace(0, 30, 100)  # Time range in days, adjust as needed

# Initialize empty arrays for DO concentration at different times
DO_concentration = np.zeros_like(time)

# Calculate DO concentration over time
for i, t in enumerate(time):
    DO_source = calculate_DO_saturation(T_wastewater)
    BOD_source = BOD_wastewater_20C * (1 - np.exp(-kd * t))
    DO_concentration[i] = DO_source - BOD_source + (BOD_source - BOD5_river) * (1 - np.exp(-kr * t))

# Calculate the critical oxygen deficit and its time
critical_oxygen_deficit = min(DO_concentration)
time_critical_deficit = time[np.argmin(DO_concentration)]

# Plot the DO profile
plt.figure(figsize=(10, 6))
plt.plot(time, DO_concentration, label='Original')

# Change the values of kd and kr
kd_doubled = 2 * kd
kr_increased = 1.5 * kr

DO_concentration_doubled_kd = np.zeros_like(time)
DO_concentration_increased_kr = np.zeros_like(time)

for i, t in enumerate(time):
    DO_source = calculate_DO_saturation(T_wastewater)
    BOD_source = BOD_wastewater_20C * (1 - np.exp(-kd_doubled * t))
    DO_concentration_doubled_kd[i] = DO_source - BOD_source + (BOD_source - BOD5_river) * (1 - np.exp(-kr * t))

    BOD_source = BOD_wastewater_20C * (1 - np.exp(-kd * t))
    DO_concentration_increased_kr[i] = DO_source - BOD_source + (BOD_source - BOD5_river) * (1 - np.exp(-kr_increased * t))

plt.plot(time, DO_concentration_doubled_kd, label='Doubled kd', linestyle='--')
plt.plot(time, DO_concentration_increased_kr, label='Increased kr', linestyle='--')

# Highlight the critical oxygen deficit
plt.scatter(time_critical_deficit, critical_oxygen_deficit, c='red', label='Critical Deficit', marker='x')

plt.xlabel('Time (days)')
plt.ylabel('Dissolved Oxygen (mg/L)')
plt.title('Dissolved Oxygen Profile over Time')
plt.grid(True)
plt.legend()
plt.show()

# Print the critical oxygen deficit and associated time
print("Critical Oxygen Deficit: {:.2f} mg/L".format(critical_oxygen_deficit))
print("Time at Critical Oxygen Deficit: {:.2f} days".format(time_critical_deficit))

```
